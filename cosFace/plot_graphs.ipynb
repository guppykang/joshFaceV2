```python
import matplotlib.pyplot as plt
from pathlib import Path 
import numpy as np
```

```python
loss_path = Path('./') / 'checkpoint' / 'trainingLog_0.txt'
loss_file = open(loss_path)
```

```python
every_other = True
losses = []
for row in loss_file:
    if every_other:
        losses.append(row.split()[5])
        every_other = False
    else:
        every_other = True
losses = losses[::20] * 10
losses
```

```python
x_axis = np.arange(len(losses)) + 1
plt.figure()
plt.plot(x_axis, losses, label='train loss')
plt.legend(loc='upper right')
plt.xlabel('iterations')
plt.ylabel('cross entropy error')
```
